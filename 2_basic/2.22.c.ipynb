```python
from sympy import *
from syscirc.funcs import u, plotcont
from syscirc.convolution import convolution_s
```

```python
def x(t:float):
    return sin(pi*t)*u(t)*u(2-t)

plotcont(x, [-1,3], 0.01)
```

```python
def h(t:float):
    return 2*u(t-1)*u(3-t)

plotcont(h, [0,4], 0.01)
```

```python
y = convolution_s(x, h)
plotcont(y, [0, 6], 0.05)
```

```python
t = Symbol("t")
ym = Piecewise((0, t<1), (1, t<3), (-1, t<5), (0, True))*2/pi*(1+cos(pi*t))
ym
```

```python
plotcont([y, ym+0.05], [0,6], 0.05)
```
