```python
from sympy import *
from syscirc.funcs import u, plotdiscr
```

```python
k, n, m = symbols('k n m', integer=True)
```

```python
w = summation((-1/2)**(k), (k, 0, n))
w.simplify()
```

```python
w_2 = 2/3*(1-(-1/2)**(n+1))
w_2
```

```python
y_1 = summation(w.subs(n, k), (k, 0, n))
y_1 = y_1.simplify()
y_1
```

```python
y_1_ = 2/3*summation(1, (k, 0, n)) + 1/3*summation((-1/2)**(k), (k, 0, n))
y_1_.simplify()
```

```python
y_1_ = 2/3*(n+1) + 1/3*(1-(-1/2)**(n+1))/(3/2)
y_1_.simplify()
```

```python
y_2 = 1/2*summation(w.subs(n,k), (k, 0, n-1))
y_2
```

```python
y = y_1 + y_2
y = y.simplify()
y
```

```python
plotdiscr(w, [-3, 4], [n])
plotdiscr(y_1, [-3, 4], [n])
plotdiscr(y_2, [-3, 4], [n])
plotdiscr(y, [-3, 4], [n])
```
