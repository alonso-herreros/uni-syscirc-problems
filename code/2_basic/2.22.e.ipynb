```python
from sympy import *
from syscirc.funcs import *
from syscirc.plot import plotcont
from syscirc.convolution import convolution_discretized
```

```python
t, τ = symbols('t τ' , real=True)
k = symbols('k', integer=True)
```

```python
def x(t): return 1 if int(abs(t)+0.5)%2==0 else -1
plotcont(x, [-3, 3], 0.05, vars=[t])
```

```python
def h(t): return 1-t if t>=0 and t<=1 else 0
plotcont(h, [-2, 2], 0.01, label="h(t)")
```

```python
y = convolution_discretized(x, h)
y(-0.5)
```

```python
plotcont([y], [-3, 3], 0.01)
```

```python
plotcont([lambda t: -1/4+t+t**2 if -2<t and t<1 else None, lambda t: 1/4+t-t**2 if -1<t and t<2 else None, y], [-3, 3], 0.01)
```

```python
def y_true(t): return -1/4+t+t**2 if -1.5<=t and t<-0.5 else 1/4+t-t**2 if -0.5<=t and t<=0.5 else y_true(t-2) if t>0.5 else y_true(t+2) if t<-1.5 else None
```

```python
plotcont([y_true], [-3, 3], 0.01)
```
