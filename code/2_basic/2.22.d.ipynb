```python
from sympy import *
from syscirc.funcs import u
from syscirc.plot import plotcont
from syscirc.convolution import convolution_s
a, b = symbols("a b")
```

```python
def x(t:float):
    return a*t + b

plotcont(x, [-1,4], 0.01, a=0.3, b=1)
```

```python
def h(t:float):
    return 2*u(t-1)*u(3-t)

plotcont(h, [0,4], 0.01)
```

```python
y = convolution_s(x, h)
plotcont(y, [0, 6], 0.05, a=0.3, b=1)
```

```python
t = Symbol("t")
ym = x
```

```python
plotcont([y-0.05, ym], [0,6], 0.05, a=0.3, b=1)
```

```python
y.subs(t, 4)
```
