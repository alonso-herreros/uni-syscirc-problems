```python
from syscirc.plot import *
```

```python
def h(n):
    if n%1 !=0: raise ValueError("h(n) is a discrete function, so n must be an integer");
    support = [0, 6]
    values = [1, 5, 10, 11, 8, 4, 1]
    return values[n-support[0]] if n >= support[0] and n <=support[1] else 0
```

```python
def h1(n):
    if n%1 !=0: raise ValueError("h1(n) is a discrete function, so n must be an integer");
    init_n = 0
    return h(n) - 2*h1(n-1) - h1(n-2) if n>=init_n else 0
```

```python
time = range(-2, 8)
print("Values: " + str.join(" ", [f"{h1(n): ^4}" for n in time]))

print("Time:   " + str.join(" ", [f"{n: ^4}" for n in time]))
```

```python
plotdiscr(h, [-1, 8])
plotdiscr(h1, [-1, 8])
```

```python
def y(n):
    if n%1 !=0: raise ValueError("y(n) is a discrete function, so n must be an integer");
    return h(n)-h(n-1)
```

```python
plotdiscr(y, [-1, 8])
```
