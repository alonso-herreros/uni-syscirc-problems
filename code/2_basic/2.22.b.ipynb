```python
from sympy import *
from syscirc.funcs import u, equivalent_test
from syscirc.plot import plotcont
from syscirc.convolution import convolution_s
a, b = symbols("a b")
```

```python
def x(t:float):
    return u(t) - 2*u(t-2) + u(t-5)

#plotcont(x, [-1,6], 0.01, a=1)
```

```python
def h(t:float):
    return exp(2*t)*u(1-t)

#plotcont(h, [-1,5], 0.01, b=1)
```

```python
y = convolution_s(x, h)
y.simplify()
```

```python
plotcont(y, [-1, 7], 0.05)
```

```python
t = Symbol("t")
ym = Piecewise((1/2*(exp(2*t)-2*exp(2*t-4)+exp(2*t-10)), t<1), (1/2*(exp(2)-2*exp(2*t-4)+exp(2*t-10)), t<3), ((1/2*(exp(2*t-10)-exp(2))), t<6), (0, True))
ym.simplify()
```

```python
plotcont([y-0.05, ym], [-1,7], 0.05, a=1, b=2)
```

```python
equivalent_test(y, ym)
```
