```python
from syscirc.funcs import *
```

```python
def x(n):
    support = [-2, 3]
    values = [1, 2, 3, 2, 2, 1] 
    return values[n-support[0]] if n >= support[0] and n <=support[1] else 0
```

```python
def y(n):
    init_n = -2
    return x(n) + 2*x(n-2) - 2*y(n-1) if n>=init_n else 0
```

```python
time = range(-4, 8)
print("Values: " + str.join(" ", [f"{y(n): ^4}" for n in time]))

print("Time:   " + str.join(" ", [f"{n: ^4}" for n in time]))
```

```python
plotdiscr(x, [-4, 6])
plotdiscr(y, [-4, 6])
```
